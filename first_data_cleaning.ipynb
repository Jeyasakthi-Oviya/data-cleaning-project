```python
dirty_data['Date']=dirty_data['Date'].str.replace("'","")
```

```python
dirty_data
```

```python
def change_date(Date):
    if pd.isnull(Date):
        return Date
    Date = str(Date)
    if Date[4] == '/' and Date[7]=='/':
        return Date
        pass
    else:
        Date= Date[:4]+'/'+Date[4:6]+'/'+Date[6:]
        return Date
       
```

```python
dirty_data['Date']=dirty_data['Date'].apply(change_date)
```

```python
dirty_data
```

```python
dirty_data['Date']=dirty_data['Date'].fillna(dirty_data['Date'].mode()[0])
```

```python
dirty_data
```

```python
dirty_data.isnull().sum()
```

```python
dirty_data['Calories']=dirty_data['Calories'].fillna(dirty_data['Calories'].mean())
dirty_data
```

```python
def duration_value(Duration):
    if Duration>100:
        return Duration/10
    else:
        return Duration
```

```python
dirty_data['Duration']=dirty_data['Duration'].apply(duration_value).astype(int)
```
